```python
import pandas as pd

df1 = pd.read_excel("Data-file-Europe-Power-Sector-2020.xlsx", sheet_name= "Data")
df2 = df1.loc[df1['Variable'] != 'Production']
df3 = df2.loc[df1['Variable'] != 'Net imports']
df4 = df3.loc[df1['Variable'] != 'Demand']
df5 = df4.loc[df4['Area'] != 'EU-27']
df6 = df5.loc[df5['Area'] != 'EU27+1']
df7 = df6.loc[df6['Variable'] != 'Wind and solar']
df7.to_csv("Pure_Data.csv")
```

```python
data = pd.read_csv("Pure_Data.csv")
data
```

```python
dt = pd.read_csv("Pure_Data.csv")
dtch = dt[['Year','Area','Variable','Share of production (%)','Change on last year (%)']]
dtch = dtch.loc[dtch['Variable'] != 'Fossil']
dtch = dtch.loc[dtch['Variable'] != 'Coal']
dtch = dtch.loc[dtch['Variable'] != 'Hard Coal']
dtch = dtch.loc[dtch['Variable'] != 'Lignite']
dtch = dtch.loc[dtch['Variable'] != 'Gas']
dtch = dtch.loc[dtch['Variable'] != 'Other fossil']
dtch = dtch.loc[dtch['Variable'] != 'Renewables']
dtch.to_csv("Renewable_Data.csv")
```

```python
dtch = pd.read_csv("Renewable_Data.csv")
dtch.fillna(0)

dtch['Share_sum(%)'] = dtch.groupby(['Year', 'Area'])['Share of production (%)'].transform(sum)
dtch['Change_sum(%)'] = dtch.groupby(['Year', 'Area'])['Change on last year (%)'].transform(sum)
dtch2 = dtch.fillna(0)
dtch2 = dtch2[['Year','Area','Share_sum(%)','Change_sum(%)']]
#dtch2.to_csv("Renewable_Data_2.csv")
```
