```python
import pandas as pd
```

```python
dt1 = pd.read_csv("Data2.csv")

dt2 = dt1.rename(columns={'Unnamed: 0':'ID'})

data = dt2[['Year','Area','Change on last year (TWh)']]

data['Total changes compared to last year (TWh)'] = data.groupby(['Year', 'Area'])['Change on last year (TWh)'].transform(sum)

data2 = data[['Year','Area','Total changes compared to last year (TWh)']]

data2.drop_duplicates()
```
